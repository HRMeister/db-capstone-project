### Join query

```python
join_query = """
            SELECT CONCAT(Customers.FirstName, " ", Customers.LastName) AS FullName, 
            Customers.PhoneNumber AS ContactInfo, 
            Orders.TotalCost AS BillAmount
            FROM Customers 
            INNER JOIN Bookings ON Customers.CustomerID = Bookings.CustomerID
            INNER JOIN Orders ON Bookings.BookingID = Orders.BookingID
            WHERE Orders.TotalCost > 60
            """
cursor.execute(join_query)
```
